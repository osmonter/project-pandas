# Data Cleaning

**Vamos a ir explorando cada columna, cuáles podemos borrar por no ser útiles, cuáles tienen info duplicada, etc.**

```python
# Hemos visto antes que especialmente Age, Time, Species tiene muchos valores nulos.
# Pero empezamos por Unnamed:22 y Unnamed:23, que prácticamente todos los valores son nulos.
```

**Columnas Unnamed:22 & Unnamed:23**

```python
# Vamos a comprobar qué datos hay en las Unnamed:22 y Unnamed:23 y después borraremos estas columnas.
```

```python
data['Unnamed: 22'].unique()
```

```python
data['Unnamed: 23'].unique()
```

```python
# No tienen ninguna información relevante, y todo el resto de valores son nulos. 
# Por tanto, eliminamos estas 2 columnas.
# Con axis = 1 hacemos referencia a que queremos borrar las columnas.

data = data.drop(['Unnamed: 22', 'Unnamed: 23'], axis=1)
```

**Columna Investigator or Source**

```python
# Hay una columna que no me resulta demasiado útil en cuanto a la información que aporta.
# Se trata de 'Investigator or Source'. 
# Parece ser la persona que escribió cada informe, o la fuente de dónde se ha sacado la info.

# Vamos a ver exactamente cuáles son los valores únicos de esta columna, y decidimos si borrarla o no.

# Aquí al hacer el método unique() no me muestra todos los valores, porque son muchas rows. 
# Haré como alternativa un subset con esta columna, y luego head(50) para ver muestra de 50 registros.

investigator_subset = data['Investigator or Source']
investigator_subset.head(50)
```

```python
# Subset me confirma que es información poco importante. Borro esta columna del dataframe.

data = data.drop(['Investigator or Source'], axis=1)
```

**Columnas Case Number / Case Number.1 / Case Number.2**

```python
# Vamos ahora con las 3 columnas de Case Number. 
# Hago un subset con ellas para verlas mejor.

casenumber_subset = data[['Case Number','Case Number.1','Case Number.2']]
casenumber_subset.head(50)
```

```python
# A simple vista, en los 50 primeros registros, parecen 3 columnas idénticas.
# Pero como son muchas rows, y 50 ejemplos serían pocos para confirmarlo, vamos a comparar primero Case Number con Case Number.1 y después Case Number con Case Number.2

# Para esta primera comparación, crearemos una nueva columna que será True si Case Number y Case Number.1 son iguales.
# Esta columna será False si son diferentes.

casenumber_subset['CN vs CN.1'] = casenumber_subset['Case Number'] == casenumber_subset['Case Number.1']
print(casenumber_subset['CN vs CN.1'])
```

```python
# Haremos lo mismo para comparar Case Number y Case Number.2

casenumber_subset['CN vs CN.2'] = casenumber_subset['Case Number'] == casenumber_subset['Case Number.2']
casenumber_subset['CN vs CN.2']
```

```python
# Vamos a contar cuántos True (es decir, repetidos) hay en ['CN vs CN.1']

casenumber_subset['CN vs CN.1'].value_counts()
```

```python
# Vamos a contar cuántos True (es decir, repetidos) hay en ['CN vs CN.2']

casenumber_subset['CN vs CN.2'].value_counts()
```

```python
# Es decir, CN vs CN.1 es True (se repite) 5979 veces, de un total de 5992. Se puede considerar duplicada.
# Por su lado, CN vs CN.2 es True (se repite) 5990 veces, de un total de 5992. Se puede considerar duplicada.

# Eliminamos por tanto de nuestro dataframe tanto Case Number.1 como Case Number.2

data = data.drop(['Case Number.1', 'Case Number.2'], axis=1)
```

**Columna Original Order**

```python
# Otra columna que no parece aportar una información relevante es "Original Order"
# Vamos a ver qué es esta columna.

data['original order'].value_counts().head(50)
```

```python
# Parece ser una especie de índice o de numeración de cada row, no parece relevante. Lo borramos.

data = data.drop(['original order'], axis=1)
```

**Columna Name**

```python
# Otra columna que no me parece importante, es la de Name. 
# No influye en nada el nombre de la persona que ha sido atacada.
# Por tanto, borro la columna de name también.

data = data.drop(['Name'], axis=1)
```

**Columnas pdf / href / href formula**

```python
# Vamos a revisar ahora las columnas, 'pdf', href' y 'href formula' porque parecen iguales o muy parecidas.
# Por el valor de cada columna, parecen documentos .pdf que guardan el informe de cada ataque.

document_subset = data[['pdf','href','href formula']]
document_subset.head(50)
```

```python
# Nuevamente, podríamos hacer como con Case Number y comparar cada registro para ver si son repetidos o no.
# Sin embargo, el objetivo de nuestro análisis es limpiar la base de datos de tal forma que nos permita sacar conclusiones o nos desvele información sobre los ataques de tiburón.

# Por tanto, en este caso creo que no es importante tener columnas que nos hagan referencia a los informes .pdf.
# Eliminaremos estas columnas, porque dan información que en nuestro análisis no es necesaria.

data = data.drop(['pdf', 'href', 'href formula'], axis=1)
```

```python
# Vamos a hacer un resumen de cómo tenemos ahora la tabla tras lo que hemos ido eliminando.

data.info()
```

**Columna Time**

```python
# Vemos que "Time" tiene una gran cantidad de nulls.

time_subset = data['Time']
time_subset.head(50)
```

**Columna Date**

```python
# Tanto con Date como con Time, me hubiera gustado poder agruparlos de alguna manera.
# Por ejemplo con Time haber separado momentos del día (Morning, Afternoon, Evening) pero no he conseguido saber hacerlo.
# Con Date, lo mismo. He intentado darles algún sentido agrupándolo por meses. Pero no me ha salido.
# Pendiente de preguntar a TA tras la correción.

# Por tanto, lo que hago con Date y Time es borrarlos como columna. Me centraré en Year de cara a situar en el tiempo los ataques.

data = data.drop(['Date','Time'], axis=1)
```

**Columna Year**

```python
# Vamos con la columna Year. 
# Convertiré todas las entradas a valor numérico, y las que no se puedan me devolverá NaN (al indicar errors='coerce')

data['Year'] = pd.to_numeric(data['Year'], errors='coerce')
data['Year'].value_counts().head(50)
```

**Columna Age**

```python
# Exploramos ahora la columna Age.
# Vemos que Age tiene "pocos" valores non-null (es decir, "muchos" nulls)
# Además está como object en lugar de como valor numérico.
# Lo tendremos que convertir también a valor numérico.

# Exploramos primero los valores de la columna Age

data['Age'].value_counts()
```

```python
# Pasamos el valor a numérico. 
# Nuevamnete, al indicar errors='coerce' insertamos NaN en aquellos valores que no se puedan convertir a número.

data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data['Age'].value_counts()
```

```python
# Vamos a ver si hay algún valor de edad desproporcionado que pueda tratarse de un error. 
# Vemos que la máxima edad es 87 años, lo cual es posible, así que damos por bueno el conjunto de edades.

data['Age'].max()
```

**Columna Species**

```python
# Vamos ahora con la columna de 'Species' que también nos indicaba que tenía muchos valores null.

data['Species'].value_counts()
```

```python
# Parece un campo donde hay información muy variada, descripciones largas, con muchas palabras
# Las especies "mejor definidas" sería White shark, Tiger shark, Bull shark, pero en el mejor de los casos se define bien 161 veces.
# Considero que 161 veces como mejor clasificación es insignificante respecto a 5992 rows que tenemos.
# Por tanto, borro esta columna de Species.

data = data.drop(['Species'], axis=1)
```

**Columna Sex**

```python
# Vamos a explorar la columna Sex

data['Sex'].value_counts()
```

```python
# Hay algunos que están mal introducidos, los vamos a cambiar por "Unknow"

data['Sex'] = data['Sex'].replace(['lli', 'N', '.', 'M '], 'Unknow')
```

```python
# Obtenemos 4835 registros que son Male (Hombres) y 585 que son Female (Mujeres)
# 5 registros son Unknow

data['Sex'].value_counts()
```

**Columna Fatal**

```python
# Exploramos ahora la columna 'Fatal (Y/N)' que indica si el ataque ha sido mortal o no.

data['Fatal (Y/N)'].value_counts()
```

```python
# Hay algunos que están mal introducidos, los corregimos.

data['Fatal (Y/N)'] = data['Fatal (Y/N)'].replace([' N', 'n', 'N '], 'N')
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].replace(['#VALUE!', 'F'], 'UNKNOWN')
```

```python
# Nos encontramos con 4325 casos NO MORTALES, con 1552 casos MORTALES, y con 96 con resultado desconocido.
data['Fatal (Y/N)'].value_counts()
```

**Columna Country**

```python
# Vamos a explorar ahora la columna Country

data['Country'].unique()
```

```python
# Veo que no todos están correctamente escritos.

# Unifico por ejemplo todos a mayúsculas

data['Country'] = data['Country'].str.upper()
```

```python
# Veo que algunas entradas acaban en signo de interrogación. Se lo quito.

data['Country'] = data['Country'].replace(['?',' '])
```

```python
# Reviso el nuevo conjunto de países tras la limpieza.

data['Country'].unique()
```

```python
# Vemos como Estados Unidos y Australia son los países que tienen más ataques. Seguidos de South Africa.
data['Country'].value_counts()
```

**Columna Location**

```python
data['Location'].value_counts().head(50)
```

```python
# Respecto a la columna de Location, aparecen muchos valores muy diversos y no aportan una gran información agrupada.
# Será una columna que borro. Nos centraremos más en Country de cara a localizar los ataques.

data = data.drop(['Location','Area'], axis=1)
```

```python
# Con la tabla que nos va quedando, considero que puede ser interesante un análisis en el que tengamos veamos
# una relación entre los países donde se producen más ataques, así como de las Activity que tienen mayor riesgo 
# y si son ataques mortales o no. Nos fijaremos en los datos que tenemos desde 1900 hasta ahora.

# Por otro lado, también es interesante fijarnos en los datos que tenemos para conocer cómo se reparten
# los ataques según Hombres / Mujeres y rangos de edades.

# Quito por tanto 'Type' e 'Injury' ya que no las usaremos en nuestros análisis.
data = data.drop(['Type', 'Injury'], axis=1)
```

```python
# Vamos a delimitar nuestro DataFrame para aquellos registros que tenemos a partir de 1900.

data = data[data.Year > 1900]
```

# Dataframe Análisis & Conclusiones

**Con la información que hemos limpiado y nos queda en nuestro dataframe, vamos a sacar algunas conclusiones.**

## Número de ataques por rango de edades

```python
# Vamos a agrupar por rangos de edades

age_labels = ['Niños (<18años)', 'Jóvenes (18-30 años)', 'Adultos (31-60)', 'Mayores (>61 años)']

cutoffs = [19,31,61,66]

data['Rango Edad'] = pd.cut(data['Age'],4 , labels=age_labels)

data.head()
```

```python
# Vamos a ver cuántos ataques de tiburón hay para cada rango de edad.

data_per_age = data.groupby(['Rango Edad']).size()
data_per_age
```

**Desde 1900 el mayor número de ataques se sitúa en Niños < 18 años y Jóvenes de 18-30 años.**

```python
# Lo vamos a mostrar como porcentaje, para que sea más significativo.

data_per_age.sum()
```

```python
data_per_age/data_per_age.sum()*100
```

**El 47% de ataques es a personas menores de 18 años, y el 40% a personas entre 18 y 30 años**
